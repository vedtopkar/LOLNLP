```python
from __future__ import division
from IPython.display import Image as PyImage
import pandas as pd
import numpy as np
import math
from PIL import Image
import requests
from StringIO import StringIO
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
PyImage(url='http://s3-ec.buzzfed.com/static/2014-05/enhanced/webdr05/23/2/anigif_enhanced-740-1400825298-3.gif')
```

```python
GIFs = pd.read_csv('data/crowdsourced_gif_emotions/gifs.csv', names=['id', 'gif_id', 'x', 'y'])
links = pd.read_csv('data/crowdsourced_gif_emotions/links.csv', names=['gif_id', 'url'])

data = GIFs.merge(links)
data['x'] = data.apply(lambda gif: (gif['x']-0.5),axis=1)
data['y'] = data.apply(lambda gif: (gif['y']-0.5),axis=1)
data.head()
```

```python
# Returns the closest emotion (assuming emotions are equally spaced) and strength (defined as distance from the origin) of an x, y pair
def get_closest_emotion(x,y):
    if (x == 0):
        x = 0.0000000000000000001 # avoid division by zero
    angle = math.atan(y/x)
    if x < 0:
        angle += math.pi
    emotions = ['pleasant', 'happy', 'elated', 'excited',
                'alert', 'active', 'tense', 'nervous', 'stressed',
                'upset', 'unpleasant', 'sad', 'unhappy', 'depressed',
                'bored', 'subdued', 'calm', 'relaxed', 'serene', 'contented']
    
    return emotions[int(round(angle/(2*math.pi/len(emotions))))]

def get_strength(x,y):
    length = math.hypot(x,y)
    return length
```

```python
data['strength'] = data.apply(lambda gif: get_strength(gif['x'],gif['y']),axis=1)
data['emotion'] = data.apply(lambda gif: get_closest_emotion(gif['x'],gif['y']),axis=1)
data.head()
data.to_csv('data/categorized_emotions.csv',encoding='utf-8')
```

```python
plt.scatter(data['x'],data['y'])
plt.xlabel("X coordinate")
plt.ylabel("Y coordinate")
plt.title("Sentiment Classification of Each GIF")
currentx = plt.xlim()
currenty = plt.ylim()
plt.plot((-1,1),(0,0), c='black', linewidth=1)
plt.plot((0,0),(-1,1), c='black', linewidth=1)
plt.xlim(currentx)
plt.ylim(currenty)
plt.show()
```

```python
pd.value_counts(data['emotion']).plot(kind="bar")
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.title('Count of Each Emotion')
plt.show()
```

```python
emotions = pd.value_counts(data['emotion']).index
emotion_strengths = []
for emotion in emotions:
    strengths = np.asarray(data[data['emotion']==emotion]['strength'])
    emotion_strengths.append(strengths)
g = sns.boxplot(emotion_strengths, names=emotions)
g.set_xticklabels(emotions,rotation='vertical')
plt.xlabel('Emotion')
plt.ylabel('Emotion Strength')
plt.title('Strength of Emotion')
plt.show()
```
