```python
import urllib2
from BeautifulSoup import BeautifulSoup
import pandas as pd
from HTMLParser import HTMLParser
parser = HTMLParser()
```

```python
# Simple example of scraping a single page for gifs

page = urllib2.urlopen('http://www.buzzfeed.com/mallorymcinnis/look-at-these-illustrated-gifs-please').read()
soup = BeautifulSoup(page)
soup.prettify()
imgs = []
texts = []
title = "blank"
for tag in soup.findAll('img'):
    if tag.has_key('src'):
        img = tag['src']
        if(img[-12:] == "_preview.gif"):
            img = img[:-12] + ".gif"
            imgs.append(img)
for tag in soup.findAll('h2'):
    text = ''.join(tag.findAll(text=True))
    text = text.replace('&#8220;','\"')
    text = text.replace('&#8217;','\'')
    texts.append(text)
for tag in soup.findAll('title'):
    title = ''.join(tag.findAll(text=True))
    title = title.replace('&#39;','\'')
dfs = []
for i in xrange(len(imgs)):
    dfs.append({'URL' : imgs[i], 'Text' : texts[i], 'Title' : title})
pd.DataFrame(dfs)
        
```

```python
# function that takes in a url and appends URL, Text, and Page Title to a list of dictionaries

def scrapepage(url):
    dfs = []
    try:
        page = urllib2.urlopen(url).read()
        soup = BeautifulSoup(page)
        soup.prettify()
        imgs = []
        texts = []
        title = "blank"
        for tag in soup.findAll('img'):
            if tag.has_key('src'):
                img = tag['src']
                if(img[-12:] == "_preview.gif"):
                    img = img[:-12] + ".gif"
                    imgs.append(img)
        for tag in soup.findAll('h2'):
            text = parser.unescape(''.join(tag.findAll(text=True)))
            texts.append(text)
        for tag in soup.findAll('title'):
            title = parser.unescape(''.join(tag.findAll(text=True)))
        for i in xrange(min(len(imgs),len(texts))):
            dfs.append({'URL' : imgs[i], 'Text' : texts[i], 'Title' : title})
        return dfs
    except:
        print (u"Could not open page " + url)
        return []
```

```python
# Make sure the function works

gifs = scrapepage('http://www.buzzfeed.com/mallorymcinnis/look-at-these-illustrated-gifs-please')
pd.DataFrame(gifs)
```

```python
# Scrape gifs from each page

urlfile = open('data/buzzfeed_gif_links.txt','r')
pages = {}
for line in urlfile:
    line = 'http://www.buzzfeed.com/' + line.replace('\n','')
    if not(line in pages):
        pages[line] = 1
urlfile_new = open('data/buzzfeed_gif_links_pt2.txt','r')
for line in urlfile_new:
    line = 'http://www.buzzfeed.com/' + line.replace('\n','')
    if not(line in pages):
        pages[line] = 1
gifs = []
for page in pages:
    gifs.append(scrapepage(page))
bf = [item for sublist in gifs for item in sublist]
pd.DataFrame(bf)
```

```python
buzzfeed_gifs = pd.DataFrame(bf)
buzzfeed_gifs.to_csv('data/buzzfeed_gifs_master.csv',encoding='utf-8')
```
