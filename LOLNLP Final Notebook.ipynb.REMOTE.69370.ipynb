## Crawling BuzzFeed

<img src='http://media.giphy.com/media/HVmfekBkrqncs/giphy.gif'/>

The first step in this process is for us to crawl BuzzFeed and collect a large list of URLs of articles that contain GIFs

````python
%matplotlib inline
import pandas as pd
import requests
import Queue
import time
````

````python
srcs = ["shawnkathleen/flying-this-holiday-season-dont-be-one-of-these-iakw"]
# This is one of many source URLs that we use to collect GIF listicles
def diff(a, b):
    """Return the different between two lists."""
    b = set(b)
    return [c for c in a if c not in b]

def parse_href(url_string):
    """Parse the URL from an HTML string containing one. If no URL, return False."""
    href_idx = url_string.find('href="')
    
    if href_idx <= 0:
        return False
    
    partial_string = url_string[href_idx+7:]
    return partial_string[0:partial_string.find('"')]

def is_gif_article(gif_article):
    """Return True if the article is a listicle containing GIFs, False otherwise."""
    
    buzzfeed_article = requests.get("http://www.buzzfeed.com/"+gif_article)
    listicle_items = buzzfeed_article.content.split('class="buzz_superlist_item')
    
    for item in listicle_items[1:-1]:
        # Check if item contains a GIF.
        if item.find('.gif') > 0:
            return True
    return False

def find_other_gif_articles(gif_article):
    """Search 'related articles' listed on input article for possible GIFs."""
    gif_articles = []
    
    buzzfeed_article = requests.get("http://www.buzzfeed.com/"+gif_article)
    related_strings = buzzfeed_article.content.split('rel:gt_act="related-link/name"')

    for related_string in related_strings[1:]:
        href = parse_href(related_string)
        if is_gif_article(href) > 0:
            gif_articles.append(href)
            
    return gif_articles

def crawl_for_gif_pages(source, limit=1000, delay=0.5):
    """Crawl Buzzfeed, starting at the source, collecting pages that contain GIFs."""
    found = []

    sources = []
    sources.append(source)

    while (len(found)<limit and len(sources)>0):
        source = sources.pop()

        new_article_candidates = find_other_gif_articles(source)
        new_articles = diff(new_article_candidates, found)

        found.extend(new_articles)
        sources.extend(new_articles)
        
        time.sleep(delay)
    
    return pd.Series(found)

# Crawl all of the sources
gif_pages = crawl_for_gif_pages(srcs)
gif_pages.to_csv('data/buzzfeed_gif_links.txt')
````

## Image Color Analysis

One aspect of the GIFs that we had Sentiment Tagged that we were very intersted in analyzing was the colors of the GIFs. To this end, we pursued a number of analysis tasks, the first of which being **color palette comparisons**.

The way this code works is that it takes a GIF URL, and then follows the following steps:

1. Store the GIF in memory
2. To spare compute time, shrink the GIF to a smaller thumbnail version (preserving color)
3. Split each frame of the GIF into a separate image in-memory
4. Merge the pixel values of all of the pixels in all of the frames to create a large array of all pixels
5. Run KMeans to find the cluster centers of the aggregated pixels
6. The centers of these clusters are the dominant colors in the colorscheme of the image

````python
from IPython.display import HTML
from __future__ import division
from IPython.display import Image as PyImage
from IPython.display import SVG
from IPython import display
import pandas as pd
import numpy as np
import math
from PIL import Image
import requests
from StringIO import StringIO
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn
````

````python
def colorscheme_from_url(url, n_colors=3):
    """
    Given a url and a given number of colors to extract
    Generate a colorscheme of n_colors colors
    """
    
    
    """
    Get the GIF from the url and save it in memory
    """
    response = requests.get(url)
    img = Image.open(StringIO(response.content))
    
    
    """
    Iterate through each frame and save it as a separate image (in memory)
    """
    n = 0
    frames = []
    while img:
        i = img.copy()
        i.thumbnail((150,150))
        frames.append(i)
        n += 1
        try:
            img.seek(n)
        except EOFError:
            break
    print 'Frames: ' + str(len(frames))
    

    """
    Get all pixels from all frames into a single 2D array
    """
    all_pixels = np.array([[100,100,100]])
    for frame in frames:
        img = frame.convert('RGB')
        all_pixels = np.vstack((np.asarray(img).reshape(-1, 3), all_pixels))

    all_pixels = all_pixels[:all_pixels.shape[0] - 1]
    
    
    """
    Run the pixels through KMeans clustering with n_colors clusters
    The centers are the primary colors
    """
    clt = KMeans(n_clusters=n_colors)
    clt.fit(all_pixels)

    primary_colors = [[int(x) for x in i] for i in clt.cluster_centers_]
    return primary_colors


def display_colorscheme(url, colors):
    """
    For use in IPython Notebooks
    Given a URL and a list of RGB colors
    Output the GIF and the colorscheme in HTML
    """
    
    html = ''
    for color in colors:
        html += '<div style="background-color: rgb({0}); padding:20px;width:20px;display:inline-block;"></div>'.format(','.join(str(int(c)) for c in color))
    
    display.display(PyImage(url=url))
    display.display(HTML(html))
````

This effectively gives us the ability to extract the primary colors in any GIF, which was very exciting. Here are examples, with each GIF and its associated colorscheme shown below it.

````python
url = 'http://media.giphy.com/media/bnfczfvZS6zni/giphy.gif'
colors = colorscheme_from_url(url, n_colors=5)
display_colorscheme(url, colors)

url = 'http://media.giphy.com/media/6F5rSKknrwPXq/giphy.gif'
colors = colorscheme_from_url(url, n_colors=5)
display_colorscheme(url, colors)

url = 'http://media.giphy.com/media/iS6w9L0bYq1Ne/giphy.gif'
colors = colorscheme_from_url(url, n_colors=5)
display_colorscheme(url, colors)
````

A natural question that we asked next was if there were any trends that we could extract from this primary color analysis. The first thing that we did was compute the average luminance of the primary colors for each GIF.

Luminance is a metric for describing the "brightness" of a color as calculated by how our eyes perceieve each of the three components of light (Red, Green, Blue). The standard equation (from here) is:

$$Luminance = \sqrt{0.241\times \text{Red}^2 + 0.691\times \text{Green}^2 + 0.068\times \text{Blue}^2}$$

````python
def compute_luminance(rgb):
    return math.sqrt( 0.241*rgb[0]**2 + 0.691*rgb[1]**2 + 0.068*rgb[2]**2 )
````

````python
def str_to_list(text):
    t = text.replace('[', '').replace(']', '').replace(' ', '').split(',')
    t = np.array([int(x) for x in t]).reshape([-1, 3])
    return t

colors = []
for i in categorized_gifs['colors']:
    colors.append(str_to_list(i))
````

### Applying Luminance Analysis to entire images

Above, we did luminance analysis on only the primary colors of each GIF. We then set out to do the same analysis on *every pixel* of each GIF. The code and analysis are below:

````python
def average_luminance_from_url(url):
    """
    Given a url and a given number of colors to extract
    Get the luminances
    """
    
    
    """
    Get the GIF from the url and save it in memory
    """
    response = requests.get(url)
    img = Image.open(StringIO(response.content))
    
    
    """
    Iterate through each frame and save it as a separate image (in memory)
    """
    n = 0
    frames = []
    while img:
        i = img.copy()
        i.thumbnail((30,30))
        frames.append(i)
        n += 1
        try:
            img.seek(n)
        except EOFError:
            break
    print 'Frames: ' + str(len(frames))
    

    """
    Get all pixels from all frames into a single 2D array
    """
    all_pixels = np.array([[100,100,100]])
    for frame in frames[::2]:
        img = frame.convert('RGB')
        all_pixels = np.vstack((np.asarray(img).reshape(-1, 3), all_pixels))

    all_pixels = all_pixels[:all_pixels.shape[0] - 1]
    
    luminances = [compute_luminance(x) for x in all_pixels]
    return luminances
````

````python
"""
Run the average_luminance_from_url function on each GIF
Aggregate the results
"""

all_image_luminances = []
urls = categorized_gifs['url']

for i,url in enumerate(urls):
    luminances = average_luminance_from_url(url)
    mean = np.mean(luminances)
    median = np.median(luminances)
    std = np.std(luminances)
    all_image_luminances.append({'url': url, 'mean': mean, 'median': median, 'std': std})
````

Another aspect of Image Analysis that we were interested in is motion detection of each GIF. We hypothesized that more active GIFs have more motiion in them. Using python motion detection packages, we tested this hypothesis.

````python
from matplotlib import pyplot as plt
from matplotlib import cm
import numpy as np
import pandas as pd
import urllib, cStringIO

from PIL import Image
from PIL import ImageChops
from scipy.stats import zscore
````

````python
def image_intensity(img):
    """
    Given an image, determine the intensity of each pixel
    """

    hist = img.histogram()
    hist = np.array(hist)

    intensity = 0

    for i, p in enumerate(hist):
        intensity += np.log(i+1)*p

    return intensity
````

````python
def motion_detection(gif_link):
    """
    Given a link to a GIF, determine the amount of motion between frames
    """

    file = cStringIO.StringIO(urllib.urlopen(gif_link).read())
    gif = Image.open(file)

    frame = 0
    motion = 0
    current_frame = gif.copy().convert('LA')

    while True: 
        try:
            previous_frame = current_frame

            gif.seek(frame+1)
            current_frame = gif.copy().convert('LA')

            motion += image_intensity(ImageChops.difference(previous_frame, current_frame))

        except EOFError:
            break

        frame += 1
    if frame > 0:
        return motion/(1.0*frame)
    return 0
````

````python
"""
Given the above functions, plot the relative motion of each GIF
"""

GIFs = pd.read_csv('data/crowdsourced_gif_emotions/gifs.csv', names=['id', 'gif_id', 'x', 'y'])
links = pd.read_csv('data/crowdsourced_gif_emotions/links.csv', names=['gif_id', 'url'])

data = GIFs.merge(links)

motion_detection_pd = pd.read_csv("data/motion_detection.csv", names=['url', 'motion'], header=0)

master = data.merge(motion_detection_pd)
master['motion'] = zscore(master['motion'])

fig = plt.figure(figsize=(18,10))
plt.title("Relative Motion in GIF", size=30)
plt.xlabel('Positivity', size=20)
plt.ylabel('Arousal', size=20)
plt.hexbin(master['x'], -1*master['y'] + 1.0, C = master['motion'], gridsize=50, cmap=cm.jet)
plt.colorbar()
plt.show()
````

## Parts of Speech Tagging
<img src='http://media.giphy.com/media/1VqmDm4AHb5S0/giphy.gif'/>

Next, we wanted to do Parts of Speech (POS) tagging on the title and text associated with each GIF in our dataset. Here, we use NLTK to do parts of speech tagging. We then use the parts of speech to analyze the content and sentiment of text.

A reference for the UPenn Treebank Project parts of speech tags can be found here: http://www.ling.upenn.edu/courses/Fall_2007/ling001/penn_treebank_pos.html

````python
import nltk
import pandas as pd
import numpy as np
````

````python
gif_data = pd.read_csv('data/buzzfeed_gifs_master.csv')
texts = gif_data['Text'].fillna('')
titles = gif_data['Title'].fillna('')
````

````python
"""
Collect all of the tags for each gif-associated text
"""
all_text_tags = []
for text in texts:
    all_text_tags.append(nltk.pos_tag(nltk.word_tokenize(text)))
    
"""
Collect all of the tags for each gif-associated title
"""
all_title_tags = []
for title in titles:
    all_title_tags.append(nltk.pos_tag(nltk.word_tokenize(title)))
````

````python
"""
Extract a certain part of speech from a list of tags
"""
pos_map = {
    'noun': 'N',
    'verb': 'V',
    'adverb': 'R',
    'adjective': 'J'
}

def get_pos(text_tags, pos):
    result = []
    c = pos_map[pos]
    for tags in text_tags:
        if tags[1][0] == c:
            result.append(tags[0])
    return result

# For example
a = all_tags[0]
get_pos(a, 'noun')
````

````python
"""
Finally, here, we grab the noun, verb, adjective, and adverb parts of speech tags for each text and title for
each GIF and save it to our DataFrame
"""

nouns = []
adverbs = []
adjectives = []
verbs = []

for tags in all_text_tags:
    nouns.append(get_pos(tags, 'noun'))
    adverbs.append(get_pos(tags, 'adverb'))
    adjectives.append(get_pos(tags, 'adjective'))
    verbs.append(get_pos(tags, 'verb'))
    
gif_data['text verbs'] = verbs
gif_data['text adverbs'] = adverbs
gif_data['text adjectives'] = adjectives
gif_data['text nouns'] = nouns
gif_data['all text tags'] = all_text_tags

nouns = []
adverbs = []
adjectives = []
verbs = []

for tags in all_title_tags:
    nouns.append(get_pos(tags, 'noun'))
    adverbs.append(get_pos(tags, 'adverb'))
    adjectives.append(get_pos(tags, 'adjective'))
    verbs.append(get_pos(tags, 'verb'))
    
gif_data['title verbs'] = verbs
gif_data['title adverbs'] = adverbs
gif_data['title adjectives'] = adjectives
gif_data['title nouns'] = nouns
gif_data['all title tags'] = all_title_tags
````

#Dictionary-Based Sentiment Analysis

What we realized from the parts of speech based sentiment analysis from above is that our dataset was too small to train an accurate sentiment analysis engine that could robustly place text into one of so many different categories. To account for this, we decided to use dictionary-based sentiment analysis so that our training data would be much larger, hopefully leading to a more accurate sentiment classifier.

The dictionary we use is from the Affective Norms for English Words (ANEW) dictionary, which can be found [here](http://personal.stevens.edu/~rchen/readings/anew.pdf). More information about this dictionary is found [here](http://csea.phhp.ufl.edu/media/anewmessage.html).

````python
from math import hypot
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import random
from IPython import display
from IPython.display import Image as PyImage
````

````python
"""
Prepare our master DataFrame
Combination of the GIF data, the links data, and the GIF emotion data
"""

gif_data = pd.read_csv('data/buzzfeed_gifs_master.csv')
GIFs = pd.read_csv('data/crowdsourced_gif_emotions/gifs.csv', names=['id', 'gif_id', 'x', 'y'])
links = pd.read_csv('data/crowdsourced_gif_emotions/links.csv', names=['gif_id', 'url'])

master = pd.merge(gif_data, links, how='right', left_on='URL', right_on='url')

master = master.merge(GIFs)
master['GIF_x'] = master.apply(lambda gif: (gif['x']-0.5),axis=1)
master['GIF_y'] = master.apply(lambda gif: (-1.0*gif['y']+0.5),axis=1)

master.head()
````

````python
"""
Read in the sentiment dictionary and calculate the sentiment means
"""

sentiment_dict = pd.read_csv("data/sentiment-dictionary.csv", index_col='Word')
sentiments = sentiment_dict[['V.Mean.Sum', 'A.Mean.Sum', 'V.SD.Sum', 'A.SD.Sum']]

vm = sentiments['V.Mean.Sum']
vm = (vm - vm.mean()) / (vm.max() - vm.min())
sentiments['valence_mean'] = vm

am = sentiments['A.Mean.Sum']
am = (am - am.mean()) / (am.max() - am.min())
sentiments['arousal_mean'] = am


def estimate_word_sentiment(word):
    """
    Given a word, estimate the valence and arousal of the word 
    (placing it on the x,y coordinate plane)
    """
    valence_mean = sentiments.ix[word]['valence_mean']
    valence_std = sentiments.ix[word]['V.SD.Sum']
    
    arousal_mean = sentiments.ix[word]['arousal_mean']
    arousal_std = sentiments.ix[word]['A.SD.Sum']
    
    return valence_mean, arousal_mean, valence_std,  arousal_std
    
def estimate_phrase_sentiment(phrase):
    """
    Given a phrase, estimate the sentiment of the phrase
    Based on the aggregation of word sentiments
    """
    
    words = phrase.split(' ')
    
    total_vm = 0.0
    total_am = 0.0
    total_vsd = 0.0
    total_asd = 0.0
    count = 0
    
    for word in words:
        if word in sentiments.index:
            vm, am, vsd,  asd = estimate_word_sentiment(word)
            total_vm += vm
            total_am += am
            total_vsd += vsd
            total_asd += asd
            count += 1
    
    if count == 0:
        return False
    
    return (total_vm/count, total_am/count, total_vsd/count, total_asd/count)

def extract_points_to_plot(text_col):
    """
    Given text, do the sentiment analysis above
    Then place the sentiment on the x,y sentiment plane.
    """
    
    xs = []
    ys = []
    
    for text in text_col:
        
        sentiment = estimate_phrase_sentiment(text)
        
        if sentiment:
            xs.append(sentiment[0])
            ys.append(sentiment[1])
        else:
            xs.append(None)
            ys.append(None)
            
    return xs, ys

xs, ys = extract_points_to_plot(master['Text'].fillna(''))
master['text_x'] = xs
master['text_y'] = ys
````

We found that this dictionary based sentiment analysis was much better at placing text into a sentiment class, especially when there are many sentiments to consider.

#Content Analysis

To further refine the results from our sentiment classifier shown above, we decided to add a filter that ensures that a GIF also has content that is relavent to the input text, not just a similar sentiment. To make sure that we took full advantage of the small amoutn of training data we have (the text and titles associated with each GIF), we used a few techniques to more easily match input text with similar content tags on GIFs. 

First, we tokenized each GIF tag to make sure that punctuation, irregular spacing, and capitalization did not hinder our ability to match a tag with content in our GIF corpus. We bolstered this by filtering out "stopwords", as defined by the NLTK corpus of English stopwords (e.g. articles or single words like "the" or "a").

Second, and more importantly, we used stemming (based on the NLTK snowball stemmer to 

````python
import nltk.corpus
import nltk.tokenize.punkt
import string
import re
````

````python
# Create a set of stopwords based on the default NLTK english stopwords
# Be sure to also include punctuation and the empty string
stopwords = nltk.corpus.stopwords.words('english')
stopwords.extend(string.punctuation)
stopwords.append('')

# Instantiate a tokenizer that tokenizes a given string
# Also instantiate the NLTK stemmer
tokenizer = nltk.tokenize.punkt.PunktWordTokenizer()
stemmer = nltk.stem.snowball.SnowballStemmer('english')
 
def check_matching_tokens(a, b):
    """
    A fuzzy matcher that sees if significant word stems match between two strings.
    Inspired by the examples here: http://bommaritollc.com/2014/06/fuzzy-match-sentences-python/
    """
    
    tokens_a = [token.lower().strip(string.punctuation) for token in tokenizer.tokenize(a) if token.lower().strip(string.punctuation) not in stopwords]
    stems_a = [stemmer.stem(token) for token in tokens_a]
    
    tokens_b = [token.lower().strip(string.punctuation) for token in tokenizer.tokenize(b) if token.lower().strip(string.punctuation) not in stopwords]
    stems_b = [stemmer.stem(token) for token in tokens_b]
    
    match = set(stems_a) & set(stems_b)
    return len(match)


def find_matching_gifs_by_content(input_string, data=data, output=True):
    """
    Given an input string and a master dataframe (data)
    Find the GIFs with matching tokenized content stems (ordered by number of matches)
    """
    
    matches = []
    
    # Iterate over each text and
    texts = data['Text'].fillna('')
    titles = data['Title'].fillna('')
    
    combined_text = [texts[i] + ' ' + titles[i] for i in range(len(data))]
    for text in combined_text:
        text = re.sub(r'[^\x00-\x7F]+',' ', text)
        n_matches = check_matching_tokens(text, input_string)
        matches.append({'Match': True if n_matches > 0 else False, 'n': n_matches })

    match_table = pd.DataFrame(matches)
    df = data.copy()
    df['Match'] = match_table['Match']
    df['N Matches'] = match_table['n']
    
    if np.sum(df['Match']) > 0:
        df = df[match_table['Match']].sort(columns='N Matches', ascending=False)
        max_n = max(df['N Matches'])
        df = df[df['N Matches'] == max_n]
        i = np.random.choice(df.index, 1)[0]
        url= df.ix[i]['url']
        t = df.ix[i]['Text']
        if output:
            print t, max_n
            display.display(display.Image(url=url))
        return df
    else:
        return []
    
find_matching_gifs_by_content('dancing')
````
